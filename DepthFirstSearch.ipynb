## 1. Create Stack class


```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        #return len(self.items)==0
        return not self.items

    def push(self,item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return (self.items[-1])

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)
```

## 2. Create a tree root node class

```python
class Node:
  def __init__(self, value, left = None, right=None):
    self.value = value
    self.left = left
    self.right = right
  def __str__(self) -> str:
    return "Node(" + str(self.value) + ")" 
```

## 3. Print the tree using pre-order traversal

```python
# pre-order
def walk_pre(tree):
  if tree is not None:
    print(tree)
    walk_pre(tree.left)
    walk_pre(tree.right)
```

## 4. Print the tree using in-order traversal

```python
# in-order
def walk_in(tree):
  if tree is not None:
    walk_in(tree.left)
    print(tree)
    walk_in(tree.right)
```

## 5. Print the tree using post traversal

```python
# post-order
def walk_post(tree):
  if tree is not None:
    walk_post(tree.left)
    walk_post(tree.right)
    print(tree)
```

##6. depth first search walk using the iteration and pre-order   

```python
def walk_iter_pre(tree, stack):
  stack.push(tree)
  while stack.size()>0:
    node = stack.pop()
    if node is not None:
      print(node)
      stack.push(node.right)
      stack.push(node.left)
```

## 7. depth first search walk using the iteration and in-order

```python
def walk_iter_in(tree, stack):
  stack.push(tree)
  node = tree
  while stack.size()>0:
    if node:
      stack.push(node)
      node = node.left
    else:
      node = stack.pop()
      print(node)
      node = node.right
      
```
